```python
import requests

query = """
SELECT ?item ?itemLabel WHERE {
  ?item rdfs:label "Albert Einstein"@en.
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
}
LIMIT 5
"""

url = "https://query.wikidata.org/sparql"

headers = {
    "Accept": "application/json",
    "User-Agent": "MyWikidataApp/1.0 (maria@example.com)"
}

response = requests.get(url, params={"query": query}, headers=headers)

data = response.json()

for item in data["results"]["bindings"]:
    print(item["item"]["value"], "-", item["itemLabel"]["value"])
```

```python
import requests

query = """
SELECT ?property ?propertyLabel ?value ?valueLabel WHERE {
  wd:Q121484467 ?prop ?value .
  ?property wikibase:directClaim ?prop .
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "en". }
}
LIMIT 50
"""

url = "https://query.wikidata.org/sparql"
headers = {
    "Accept": "application/json",
    "User-Agent": "MyWikidataApp/1.0 (maria@example.com)"
}

response = requests.get(url, params={"query": query}, headers=headers)
data = response.json()

for row in data["results"]["bindings"]:
    print(row["propertyLabel"]["value"], ":", row.get("valueLabel", {}).get("value"))
```

```python
import requests

query = """
SELECT ?wiki ?article WHERE {
  VALUES ?item { wd:Q121484467 }
  ?article schema:about ?item ;
           schema:isPartOf ?wiki .
}
"""

url = "https://query.wikidata.org/sparql"
headers = {
    "Accept": "application/json",
    "User-Agent": "MyWikidataApp/1.0 (maria@example.com)"
}

response = requests.get(url, params={"query": query}, headers=headers)
data = response.json()

for row in data["results"]["bindings"]:
    print(row["wiki"]["value"], "→", row["article"]["value"])
```
